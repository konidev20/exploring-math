```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def c(n):
    if n == 2:
        return 1
    if n % 2 == 0:
        return n // 2
    else:
        return 3 * n + 1
```

```python
n = 10000
```

```python
def run(n):
    edges = []
    max_num = n
    steps = 0
    r = n
    while r != 1:
        r = c(r)
        if r > max_num:
            max_num = r
        steps = steps + 1
    return (steps, max_num)
```

```python
xpoints = np.array([i for i in range (2, n)])
spoints = np.array([])
mpoints = np.array([])

for i in range (2, n):
    x, y = run(i)
    spoints = np.append(spoints, x)
    mpoints = np.append(mpoints, y)
```

```python
f, (a1) = plt.subplots(1,1)
f.set_figwidth(20)
f.set_figheight(10)
a1.scatter(xpoints, spoints)
a1.set_xlabel('n')
a1.set_ylabel('steps')
```

```python
f2, (a2) = plt.subplots(1,1)
f2.set_figwidth(40)
f2.set_figheight(40)
a2.scatter(xpoints, mpoints)
a2.set_ylim([0, n])
a2.set_xlabel('n')
a2.set_ylabel('max_steps')
```
